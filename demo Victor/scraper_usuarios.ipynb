Va a ser la idea correcta sacar una muestra uniforme de usuarios a travez de cada rango, 1000 por el rango(1, 865000)
esta muestra ira de 1 a 17300000, con un total de 200000 usuarios

```python
import pandas as pd
import requests
import numpy as np
import json
from bs4 import BeautifulSoup as bs
import random
import time


def extraer_ids(archivo: str, rango_min, rango_max):
    try:
        with open(archivo, "r") as file:
            dict_usuarios = json.load(file)
    except FileNotFoundError:
        dict_usuarios = {}
    i = len(dict_usuarios)
    while i<1000:
        id_user = random.randint(rango_min, rango_max)
        if str(id_user) in dict_usuarios or id_user in dict_usuarios:
            time.sleep(6)
            continue
        sitio_user = requests.get(
            url="https://myanimelist.net/comments.php",
            params={"id": id_user}
        )
        if sitio_user.status_code == 429: # Demasiadas peticiones
            print("Demasiadas peticiones")
            time.sleep(120)
            continue
        elif sitio_user.status_code != 200: # Error desconocido
            print("Error desconocido")
            print(sitio_user.status_code)
            time.sleep(6)
            continue
        titulo = bs(sitio_user.content).find_all("title")[0].contents[0].strip()
        if titulo == '404 Not Found - MyAnimeList.net': # El usuario no existe
            print("Error 404")
            time.sleep(6)
            continue
        username = titulo[:-29]
        dict_usuarios[id_user] = username
        i += 1
        time.sleep(6)
    json.dump(dict_usuarios, open(archivo, "w"), indent=2)
```

```python
for i in range(2, 10):
    extraer_ids(f"usuarios_{i}.json", 865000 * (i-1) + 1, 865000 * (i))
```

```python
for i in range(10, 21):
    extraer_ids(f"usuarios_{i}.json", 865000 * (i-1) + 1, 865000 * (i))
```
