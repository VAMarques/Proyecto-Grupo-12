# El Gran Scraper

```python
### Headers para la animelist.
headers = {
    "accept": "application/json",
    "X-MAL-CLIENT-ID": "dba80244ba88ffa1c4914806e9ccb3e4"
}

### Parametros para obtener la animelist del usuario.
animelist_params = {
    "fields": "list_status",
    "sort": "list_score",
    "limit": "1000"
}

# gran_df = pd.DataFrame()
i = 0
while i<18:
    id_user = random.randint(1, 17300000)
    sitio_user = requests.get(
        url="https://myanimelist.net/comments.php",
        params={"id": id_user}
    )
    titulo = bs(sitio_user.content).find_all("title")[0].contents[0].strip()
    if titulo == '404 Not Found - MyAnimeList.net': # El usuario no existe
        continue
    username = titulo[:-29]
    user_animelist = requests.get(
        url="https://api.myanimelist.net/v2/users/"+username+"/animelist",
        headers=headers,
        params=animelist_params
    )
    if user_animelist.status_code == 403: # La Animelist del usuario es privada.
        continue
    user_json = user_animelist.json()["data"]
    if not user_json: # La Animelist es vacia.
        continue
    user_df = pd.json_normalize(user_json)
    user_df = user_df[["node.id", "list_status.score"]]
    id_combinada = f"{id_user}:{username}"
    user_df.rename(columns={"node.id":"anime_id", "list_status.score":id_combinada}, inplace=True)
    user_df.set_index("anime_id", drop=True, inplace=True)
    gran_df = pd.concat([gran_df, user_df], axis=1)
    i+=1 # Ciclo exitoso
```
